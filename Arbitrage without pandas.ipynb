# Crypto assingment

```python
import requests as r, matplotlib.pyplot as plt, numpy as np 
def crypto_defrence(crypto='btc', fiat='usd'):
    url = 'https://api.cryptonator.com/api/full/{}-{}'.format(crypto,fiat)
    json = r.get(url).json()['ticker']['markets']
    market, volume, price = [], [], []
    for i in range(len(json)): market.append(json[i]['market']); volume.append(json[i]['volume']); price.append(float(json[i]['price']))
    plt.figure(figsize=(15,8)), plt.scatter(price, volume, s=150)
    for i, txt in enumerate(market): plt.annotate(txt, (price[i],volume[i]), size= 25) 
    plt.title('crypto currency rates Graph', size=30); plt.xlabel('Rates in '+str(fiat).upper(), size=20); plt.ylabel('currencies volumes', size = 40)
    print('minimum price at:\t',market[(price.index(min(price)))],min(price) ,'\nmaximum price at:\t',market[(price.index(max(price)))],min(price), '\ndefrence :\t\t',int((max(price) - min(price))), fiat, '\nProfit ratio: \t\t', round((max(price)-min(price))/min(price),4))
    plt.show()
print(crypto_defrence())
```

```python
crypto_defrence('eth', 'eur')
```
